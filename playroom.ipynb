```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import pandas as pd
```

```python
DIR = "E:/Kuzushiji/Images/"
train_path = DIR+"train"
validate_path = DIR+"validate"

df = pd.read_csv("train.csv")
df.set_index('image_id', inplace = True)

translations = pd.read_csv("unicode_translation.csv")
translations.set_index('Unicode', inplace = True)

def img_processor(path, labels):
    img = cv2.imread(path)
    training_data = list()
    for i in labels:
        X = int(i[1])
        Y = int(i[2])
        width = int(i[3])
        height = int(i[4])
        training_data.append(img[Y:Y+height,X:X+width, :])
    training_data = np.array(training_data)
    return training_data

def data_creation(data_string):
    features = list()
    data = data_string.split()
    steps = int(len(data)/5)
    for i in range(1,steps):
        features.append(data[:5])
        data = data[5:]
        i+=1
    return features

def label_creator(features):
    labels = list()
    for i in features:
        labels.append(i[0])
    labels = np.array(labels)
    return labels
    

def Batch_generator(DIR):
    while(True):
        for img in os.listdir(DIR):
            ind = img[:-4]
            data_string = df.loc[ind].labels
            features = data_creation(data_string)
            X_train = img_processor(os.path.join(DIR,img), features)
            y_train = label_creator(features)
            yield(X_train,y_train)
```

```python
gen = Batch_generator(train_path)
```

```python
(X_train,y_train) = next(gen)
```
